```python
#Paquetes o modulos que se utilizarán para esta práctica
import random as rd #importar biblioteca random para generar un numero alternativo
from openpyxl import load_workbook #esto nos permitira modificar las celdas del excel
import pandas as pd #esto nos permitirá visualizar y modificar el excel
import matplotlib.pyplot as plt #esto nos permitirá crear las visualizaciones de la opción 3

#funcion utilizada para comprobar que las respuestas del usuario se encuentran dentro de las proporcionadas por el usuario
def validez(opciones_validas=(1,2,3,4)):
   while elegir_opcion not in opciones_validas:
      print("Opción no válida")
      break


#funcion utilizada para visualizar el resumen de las partidas
def mostrar_datos():
   datos=pd.read_excel("C:\\practicaPythonbasico\\estadisticas_juego.xlsx") #cargamos el archivo excel mediante una variable "datos"
   print("\n\tResumen de partidas") #imprimimos un encabezado para nuestra tabla
   print(datos.head())              #esta función muestra las primeras filas de nuestro tabla


#elegir modo de juego
while True:
  elegir_opcion=int(input("\n1. Jugar\n"+"2. Mostrar resumen partidas\n"
                          +"3. Mostrar gráficas partidas\n"+"Introduzca una opcion: "))

  validez() #comprobamos que la respuesta se encuentre entre las opciones

  #Opcion 1:jugar
  if elegir_opcion==1:

    #modo de dificultad
    elegir_opcion=int(input("\nEscoja el nivel de dificultad\n"+"1. Fácil\n"+"2. Medio\n"+"3. Difícil\n"+"Introduzca una opción: "))

    validez() #comprobamos que la respuesta se encuentre entre las opciones

    #dependiendo del nivel de dificultad el contador empezará en 15, 10 o 5
    if elegir_opcion==1:
        print("\nJugando en modo FÁCIL\n") #indicamos el modo de juega al usuario
        contador=15

    elif elegir_opcion==2:
        print("\nJugando en modo MEDIO\n")
        contador=10

    elif elegir_opcion==3:
        print("\nJugando en modo DIFÍCIL\n")
        contador=5

    numero_correcto=rd.randint(0,1000) #generamos un numero aleatorio entre 0 y 1000 y lo guardamos en una variable

    frase="Introduzca un número entre 0 y 1000: "

    respuesta=int(input("\n* Te quedan "+ str(contador)+" intentos\n"+frase))

    while respuesta!=numero_correcto and contador!=0: #mientras que la respuesta no sea igual al numero creado de forma aleatoria, se ira restando 1 al contador,hata que llegue a 0
        contador-=1
        if respuesta<numero_correcto: #si la respuesta es menor que el numero generado, se indicará que el numero es mayor y viceversa
          print("El numero a adivinar es MAYOR")
        else:
          print("El número a adivinar es MENOR")

        respuesta=int(input("* Te quedan "+ str(contador)+" intentos\n"+frase)) #mientras el bucle continue se seguirá imprimiendo el input

        if elegir_opcion==1 and contador==1 and numero_correcto!=respuesta:          #dependiendo de la dificultad seleccionada...
          print("\nLo siento has perdido")
          print("El numero correcto era: ", numero_correcto)                        #se indica que número se había generado
          docExcel=load_workbook("C:\\practicaPythonbasico\\estadisticas_juego.xlsx")    #se carga el fichero
          hoja= docExcel["Resultados del juego"]                                    #se carga la hoja del fichero
          docExcel["Resultados del juego"]["C2"].value+=1                           #...se sumará una derrota a la celda correspondiente
          docExcel.save("C:\\practicaPythonbasico\\estadisticas_juego.xlsx")             #tras ello se guardan los cambios en el excel
          break

        elif elegir_opcion==2 and contador==1 and numero_correcto!=respuesta:
          print("\nLo siento has perdido")
          print("El numero correcto era: ", numero_correcto)
          docExcel=load_workbook("C:\\practicaPythonbasico\\estadisticas_juego.xlsx")
          hoja= docExcel["Resultados del juego"]
          docExcel["Resultados del juego"]["C3"].value+=1
          docExcel.save("C:\\practicaPythonbasico\\estadisticas_juego.xlsx")
          break

        elif elegir_opcion==3 and contador==1 and numero_correcto!=respuesta:
          print("\nLo siento has perdido")
          print("El numero correcto era: ", numero_correcto)
          docExcel=load_workbook("C:\\practicaPythonbasico\\estadisticas_juego.xlsx")
          hoja= docExcel["Resultados del juego"]
          docExcel["Resultados del juego"]["C4"].value+=1
          docExcel.save("C:\\practicaPythonbasico\\estadisticas_juego.xlsx")
          break

    #en caso de que la respuesta sea igual al numero generado, el usuario habrá ganado
    if elegir_opcion==1 and respuesta==numero_correcto:                      #a diferencia de los condicionales de arriba, aqui no se necesita...
        print("\nEnhorabuena has ganado")                                      #...que el contador llegue a uno para que se anote la victoria
        docExcel=load_workbook("C:\\practicaPythonbasico\\estadisticas_juego.xlsx")
        hoja= docExcel["Resultados del juego"]                                 #dependiendo de la dificultad...
        docExcel["Resultados del juego"]["B2"].value+=1                         #...se sumará una victoria a la celda correspondiente
        docExcel.save("C:\\practicaPythonbasico\\estadisticas_juego.xlsx")           #se guardarán los cambios en el excel

    elif elegir_opcion==2 and respuesta==numero_correcto:
        print("\nEnhorabuena has ganado")
        docExcel=load_workbook("C:\\practicaPythonbasico\\estadisticas_juego.xlsx")
        hoja= docExcel["Resultados del juego"]
        docExcel["Resultados del juego"]["B3"].value+=1
        docExcel.save("C:\\practicaPythonbasico\\estadisticas_juego.xlsx")

    elif elegir_opcion==3 and respuesta==numero_correcto:
        print("\nEnhorabuena has ganado")
        docExcel=load_workbook("C:\\practicaPythonbasico\\estadisticas_juego.xlsx")
        hoja= docExcel["Resultados del juego"]
        docExcel["Resultados del juego"]["B4"].value+=1
        docExcel.save("C:\\practicaPythonbasico\\estadisticas_juego.xlsx")

  #Opcion 2:Mostrar resumen de partidas
  elif elegir_opcion==2:

    mostrar_datos() #utilizamos la función anteriormente creada para leer los datos

  #Opcion 3:Mostrar resumen de partidas y graficas de cada modo de juego
  elif elegir_opcion==3:
    mostrar_datos()

    docExcel=load_workbook("C:\\practicaPythonbasico\\estadisticas_juego.xlsx")
    hoja= docExcel["Resultados del juego"]

    #Grafico modo facil
    plt.figure(figsize=(15, 5))#aumentamos el tamaño predeterminado de los graficos ya que la funcion subplots_adjust los estrechará
    ejex=[hoja["B2"].value,hoja["C2"].value]
    ejey=["Victorias","Derrotas"]
    plt.subplot(1,3,1) #con esta función podremos hacer que los 3 graficos se muestren en el mismo espacio
                        #Primer numero:numero de filas, segundo numero:numero de columnas, tercer numero:la posicion de nuestro grafico
    plt.bar(ejey,ejex, color="blue")
    plt.title("PARTIDAS EN MODO FÁCIL") #titulo del gráfico

    #Grafico modo medio
    ejex=[hoja["B3"].value,hoja["C3"].value]
    ejey=["Victorias","Derrotas"]
    plt.subplot(1,3,2)
    plt.bar(ejey, ejex, color="yellow")
    plt.title("PARTIDAS EN MODO MEDIO")

    #Grafico modo dificil
    ejex=[hoja["B4"].value,hoja["C4"].value]
    ejey=["Victorias","Derrotas"]
    plt.subplot(1,3,3)
    plt.title("PARTIDAS EN MODO DIFÍCIL")
    plt.bar(ejey, ejex, color="red")
    plt.subplots_adjust(wspace=0.7) #con esta función aumentamos el espacio entre los graficos para que no queden tan juntos
    plt.show()

  elif elegir_opcion==4:
    break
```
